```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [3,4,5,6,7]
y = [41,21,31,21,22]
z = [31,21,11,25,34]
avg_t = [24,31,22,12,19]
```

```python
plt.plot(x,avg_t)
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather")
plt.plot(x,avg_t)
plt.plot(x,y)
plt.plot(x,z)
```

```python
plt.xlabel("days")
plt.ylabel("temp")
plt.title("weather")
plt.plot(x,avg_t,label='max')
plt.plot(x,y,label='min')
plt.plot(x,z,label='avg')

plt.legend()
```
