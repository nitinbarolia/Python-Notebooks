```python
blood_sugar_men = [113, 85, 90, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 89, 95, 100, 125, 99, 79, 139, 100, 141]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'])
```

```python
blood_sugar_men = [113, 85, 90, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 89, 95, 100, 125, 99, 79, 139, 100, 141]

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
        label=['men','women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 89, 95, 100, 125, 99, 79, 139, 100, 141]

plt.xlabel('sugar range')
plt.ylabel('total no of petients')
plt.title('blodd sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
        label=['men','women'])
plt.legend()
```

```python
blood_sugar_men = [113, 85, 90, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 89, 95, 100, 125, 99, 79, 139, 100, 141]

plt.xlabel('sugar range')
plt.ylabel('total no of petients')
plt.title('blodd sugar analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],
        label=['men','women'],orientation='horizontal')
plt.legend()
```
